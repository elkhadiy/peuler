```python
from numba import jit
import numpy as np

primes = np.load('utils/primes_1M.npz')['primes']

@jit
def pwrfactor(N, f):
    n = 1
    while N % f**n == 0:
        n += 1
    return n-1

@jit
def pwrfactors(N, F):
    return np.array([ pwrfactor(N, f) for f in F ])

@jit
def factors(n):
    p = primes[primes <= n]
    f = np.column_stack((p, pwrfactors(n, p).flatten()))
    return f[f[:,1] > 0]

@jit
def trinum(n):
    return np.sum(np.arange(n+1))

@jit
def divisor_cnt(n):
    return np.prod(factors(n)[:,1] + 1)

@jit
def sol(i=2, N=5):
    while divisor_cnt(trinum(i)) < N:
        i += 1
    return i
```

```python
from numba import njit

@njit
def collatz(n):
    return (n % 2) * (3*n + 1) + ((n+1) % 2) * (n / 2)

@njit
def collatz_chain_len(n):
    r = collatz(n)
    cnt = 2
    while r > 1:
        r = collatz(r)
        cnt += 1
    return cnt
```

```python
def get_neighbors(pt):
    i, j = pt
    n = np.array([
                        [i - 1, j    ], 
        [i    , j - 1],                 [i    , j + 1],
                        [i + 1, j    ]
    ])

    return n[ (0 <= n[:,0]) & (n[:,0] < 20) & (0 <= n[:,1]) & (n[:,1] < 20) ].tolist()

def is_neighbors(p1, p2):
    return (p1 in get_neighbors(p2)) & (p2 in get_neighbors(p1)) & (p1 != p2)
```

```python
def nb_paths_m(N):
    N += 1

    coords = np.array([ (i, j) for i in range(N) for j in range(N) ]).tolist()

    M = np.array([
        [ int(is_neighbors(p2, p1)) for p1 in coords ]
        for p2 in coords
    ], dtype='uint64')

    return matrix_power(M, (N-1)*2)[0, -1]
```

```python
from __future__ import annotations
from dataclasses import dataclass

@dataclass
class Node:
    value: int
    sum_best_path: int = 0
    lparent: Node = None
    rparent: Node = None
    lchild: Node = None
    rchild: Node = None

    def __add__(self, other):
        return self.value + other
    def __radd__(self, other):
        return self + other
```

```python
def load_nodes(df):
    prev_nodes = []
    for line in df:
        nodes = [ Node(int(v)) for v in line.strip().split()]
        for i, node in enumerate(nodes):
            if i > 0:
                node.lparent = prev_nodes[i-1]
                prev_nodes[i-1].rchild = node
            if i < len(prev_nodes):
                node.rparent = prev_nodes[i]
                prev_nodes[i].lchild = node
        prev_nodes = nodes
    return prev_nodes
```

```python
def best_path(node):
    if node is None:
        return []
    if sum(best_path(node.lparent)) > sum(best_path(node.rparent)):
        return best_path(node.lparent) + [node]
    else:
        return best_path(node.rparent) + [node]

def best_path_at(line):
    return max([sum(best_path(node)) for node in line])
```

```python
from numba import jit

@jit(forceobj=True)
def sum_best_path(node):
    if node is None:
        return 0
    if node.sum_best_path > 0:
        return node.sum_best_path
    node.sum_best_path = max(sum_best_path(node.lparent), sum_best_path(node.rparent)) + node
    return node.sum_best_path

def best_path_at(line):
    return max([sum_best_path(node) for node in line])
```

```python
numbers_10000 = np.arange(2, 10_000)
amicable_numbers_10000 = np.array([is_amicable(i) for i in numbers_10000])
```

```python
from functools import cache

@cache
def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n-1) + fibo(n-2)
```

```python
i = 12
while len(digits2arr(fibo(i))) < 1000:
    i += 1
```
